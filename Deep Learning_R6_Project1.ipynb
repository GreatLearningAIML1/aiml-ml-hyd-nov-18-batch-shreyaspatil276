```python
import numpy as np 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense
import tensorflow as tf
from keras import optimizers
```

```python
data=pd.read_csv('bank.csv')
```

```python
data.head()
```

```python
data.drop(['RowNumber','CustomerId','Surname'],inplace=True,axis=1)
```

```python
data.shape
```

```python
data.dtypes
```

```python
data['Geography'].value_counts()
```

```python
#Convert categorical variables using one hot encoding
ohe_df=pd.get_dummies(data['Geography'])
data['France']=ohe_df['France']
data['Germany']=ohe_df['Germany']
data['Spain']=ohe_df['Spain']
```

```python
data.drop('Geography',axis=1,inplace=True)
```

```python
data.head()
```

```python
data['Gender'].replace(['Female','Male'],[0,1],inplace=True)
data.head()
```

```python
X=data.drop(['Exited'],axis=1)
Y=data['Exited']
```

```python
#Train test split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=1)
```

```python
Y_train=tf.keras.utils.to_categorical(Y_train)
Y_test=tf.keras.utils.to_categorical(Y_test)
```

```python
#Standardizaion
sc=StandardScaler()
X_train_Std=pd.DataFrame(sc.fit_transform(X_train),columns=X_train.columns)
X_test_Std=pd.DataFrame(sc.fit_transform(X_test),columns=X_test.columns)
```

```python
# Building and optimizing model
model = Sequential()
#specifying activation functions
model.add(Dense(10, input_dim=12, init='uniform', activation='relu'))
model.add(Dense(8, init='uniform', activation='relu'))
model.add(Dense(6, init='uniform', activation='relu'))
model.add(Dense(2, init='uniform', activation='softmax'))
# Compile model
adam=optimizers.adam(lr=0.001)
model.compile(loss='binary_crossentropy', optimizer=adam, metrics=['accuracy'])
# Fit the model
model.fit(X_train_Std, Y_train, nb_epoch=100, batch_size=10,  verbose=2)
```

```python
#Predict
predicted=model.predict(X_test_Std)
```

```python
#0.5 Threshold
result=np.where(predicted>=0.5)[1]
```

```python
np.bincount(result)
```

```python
#Accuracy and Confusion matirx
model.evaluate(X_test_Std,Y_test)
```

```python
print("Accuracy is", model.evaluate(X_test_Std,Y_test)[1]*100)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
confusion_matrix(np.argmax(Y_test,axis=1),result)
```

```python
accuracy_score(np.argmax(Y_test,axis=1),result)
```
