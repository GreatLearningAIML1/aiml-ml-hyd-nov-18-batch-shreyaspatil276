<h3>Write a function with name â€œGoal_diff_countâ€ which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points)

```python
'''Arguments:This function takes Dataframe having GoalsFor,GoalsAgainst,Team and Pos column as argument
   Return:This function returns 5 values 
   1.A Dataframe with Position,Team Name and Goal Difference
   2.Name of team with maximum goal difference
   3.Name of team with minimum goal difference
   4.Value of maximum goal difference
   5.Value of minimum goal difference'''
```

```python
def Goal_diff_count(df):
    #Convert string to float
    df['GoalsFor']=df['GoalsFor'].astype(float)
    df['GoalsAgainst']=df['GoalsAgainst'].astype(float)
    
    #Calculate the goal difference and store it in a Dataframe
    Goal_Diff=pd.DataFrame(data=np.array([Laliga['Team'],Laliga.GoalsFor-Laliga.GoalsAgainst]).T,index=Laliga['Pos'],columns=['Team','GoalDifference'])
    
    #Get the name of team having maximum Goal Difference
    Team_Goal_Diff_max=Goal_Diff.sort_values('GoalDifference',ascending=False).iloc[0,0]
    
    #Get the name of team having minimum Goal Difference
    Team_Goal_Diff_min=Goal_Diff.sort_values('GoalDifference').iloc[0,0]
    
    #Return the caclculated values along with maximum and minimum values
    return Goal_Diff,Team_Goal_Diff_max,Team_Goal_Diff_min,Goal_Diff['GoalDifference'].max(),Goal_Diff['GoalDifference'].min()
```

```python
gd,team_gd_max,team_gd_min,gd_max,gd_min=Goal_diff_count(Laliga)
print(gd,"\n",team_gd_max,"\n",team_gd_min,"\n",gd_max,"\n",gd_min)
```

<h3>Q5.Create a new column with name â€œWinning Percentâ€ and append it to the data set (5 points)
Percentage of Winning = (GamesWon / GamesPlayed)*100

If there are any numerical error, replace it with 0%

Print the top 5 teams which has the highest Winning percentage

```python
#This function calculates winning percentage and in case of numerical error,it retutns 0
def CalculateWinningPercentage(df):
    df['GamesWon']=df['GamesWon'].astype(float)
    df['GamesPlayed']=df['GamesPlayed'].astype(float)
    try:
        Winning = (df['GamesWon'] / df['GamesPlayed'])*100
    except:
        Winning = 0
    return Winning
```

```python
#Add a new column to dataset
Laliga['Winning Percent']=CalculateWinningPercentage(Laliga)
Laliga
```

```python
Laliga.loc[Laliga['Winning Percent'].sort_values(ascending=False)[0:5].index][['Team','Winning Percent']]
```
